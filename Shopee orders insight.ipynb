### LOADING LIBRARY


```python
import pandas as pd
import matplotlib.pyplot as plt
```

Reading the input file

```python
df = pd.read_csv('/order_details_1.csv')
df.head() # 5 first records
```

```python
df.describe()
```

Find the product which have lowest price


```python
list_of_product = df.to_dict(orient = 'records')
min_price_product = min(list_of_product, key = lambda x:x['Giá'])
min_price_product = pd.DataFrame([min_price_product])
min_price_product
```

Find the product which have highest price


```python
max_price_product = max(list_of_product, key = lambda x:x['Giá'])
max_price_product = pd.DataFrame([max_price_product])
max_price_product
```

Money has been spent


```python
new_info = df.dropna(subset=['Trạng thái đơn hàng'])
print(type(new_info))
new_info = new_info.to_dict(orient = 'records')
```

```python
print(sum(x['Giá'] for x in new_info))
```

### DATA VISUALIZATION ###

BAR CHART

```python
na_count1 = df['Trạng thái đơn hàng'].count()
na_count1
```

```python
na_count = df['Trạng thái đơn hàng'].isnull().sum()
non_na_count = df['Trạng thái đơn hàng'].notnull().sum()

# Create the bar chart
plt.bar(['Hủy đơn', 'Hoàn thành đơn'], [na_count, non_na_count])
plt.title('Số lượng hoàn thành và hủy đơn')
plt.xlabel('Trạng thái đơn hàng')
plt.ylabel('Số lượng')
plt.tight_layout()
plt.show()
```

PIE CHART

```python
shop_list = df['Tên shop'].to_list()
print(shop_list)
shop_counts = pd.Series(shop_list).value_counts()
print(shop_counts)
# labels = shop_list
fig,ax = plt.subplots()
ax.pie(shop_counts, labels = shop_counts.index,autopct='%1.1f%%')
plt.axis('equal')
```
